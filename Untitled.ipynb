```python
import requests
import pandas as pd
import json
```

```python
url = 'https://api.squiggle.com.au/?q=games&year=2023'
data = requests.get(url).json()['games']
df = pd.DataFrame.from_records(data)
```

```python
for col in df.columns:
    if df[col].dtype == 'float64':
        df[col] = df[col].fillna(-1)
        df[col] = df[col].astype(int)
        df[col] = df[col].astype(str)
        df[col] = df[col].replace('-1', '')
```

```python
df['hteamid'] = df['hteamid'].fillna(-1)
```

```python
df['hteamid'] = df['hteamid'].astype(int)
df['hteamid'] = df['hteamid'].astype(str)
df['hteamid'] = df['hteamid'].replace('-1', '')
```

```python
from sqlalchemy import create_engine
import pandas as pd
```

```python
engine = create_engine('sqlite:///db.sqlite3', echo=True)
sqlite_connection = engine.connect()
```

```python
sql = "select * from tipping_tippings where status='completed'"
```

```python
tf = pd.read_sql(sql, engine)
```

```python
tf.shape
```

```python
tf
```

```python
df = tf.groupby(['email', 'first_name', 'last_name'])[['tips', 'margin_diff', 'round']].agg({'tips': 'sum', 'margin_diff': 'sum', 'round': pd.Series.nunique})
```

```python
df = df.reset_index()
```

```python
df
```

```python
df.columns = ['email', 'first_name', 'last_name', 'total_tips', 'total_margin', 'rounds_tipped']
```

```python
df['avg_per_round'] = df['total_tips'] / df['rounds_tipped']
```

```python
tf1 = tf.loc[tf['round']==12]
```

```python
df2 = tf1.groupby('email')[['tips', 'margin_diff']].sum().reset_index()
```

```python
df2.columns = ['email', 'last_round_tips', 'last_round_margin']
```

```python
df = df.merge(df2, on='email', how='left')
```

```python
df['full_name'] = df['first_name'] + ' ' + df['last_name']
```

```python
t1 = df.reset_index()
```
